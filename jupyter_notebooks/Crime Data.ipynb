```python
import pandas as pd
import numpy as np
```

```python
crimes = pd.read_csv("Crimes_-_2001_to_present.csv")
crimes
```

# Reported crimes per Community Area by Primary Type 

```python
crime_org= pd.DataFrame(crimes.pivot_table(crimes,index=["Community Area", "Primary Type"]))
crime_org
```

```python

crime_data = pd.DataFrame(crimes.pivot_table(values=["ID"], index=["Community Area", "Primary Type", "Date"], aggfunc=np.sum))
crime_data
```

# Number of Reported Crimes per Commuity Area 

```python
#1) group by community area
#2) get numbers of crime type per community area

crime_area= pd.DataFrame(crimes.groupby("Community Area")["ID"].count())
crime_area
```

# Number of Reported Crimes per Primary Type

```python
crime_type= pd.DataFrame(crimes.groupby("Primary Type")["ID"].count())
crime_type
```
