```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Final Tables through 2016 - Race & Ethnicity.csv').fillna(0)
```

```python
df
```

```python
areas = []
for i in range(0, len(df)):
    if df.iloc[i]['Race and Ethnicity in Chicago Community Areas'] == 0:
        areas.append(df.iloc[i+1]['Race and Ethnicity in Chicago Community Areas'].strip("1234567890. "))

    
areas
```

```python
thirty = []
fourty = []
fifty = []
sixty = []
seventy = []
eighty = []
ninety = []
thousand = []
ten = []

for i in range(0, len(df)):
    if df.iloc[i]['Race and Ethnicity in Chicago Community Areas'] == 'Total':
        thirty.append(df.iloc[i][1])
        fourty.append(df.iloc[i][2])
        fifty.append(df.iloc[i][3])
        sixty.append(df.iloc[i][4])
        seventy.append(df.iloc[i][5])
        eighty.append(df.iloc[i][6])
        ninety.append(df.iloc[i][7])
        thousand.append(df.iloc[i][8])
        ten.append(df.iloc[i][9])
        
ten
```

```python
df_clean = pd.DataFrame({
    'Community':areas,
    '1930':thirty,
    '1940':fourty,
    '1950':fifty,
    '1960':sixty,
    '1970':seventy,
    '1980':eighty,
    '1990':ninety,
    '2000':thousand,
    '2010':ten
})
```

```python
df_clean["ID"] = df_clean.index + 1
df_clean = df_clean.set_index("ID")
df_clean
```

```python
pop_history = df_clean.drop('Community', axis=1)
pop_history.to_csv('pop_history.csv')
```

```python
comm_names = df_clean.drop(['1930','1940','1950','1960','1970','1980','1990','2000','2010'], axis=1)
comm_names.to_csv('comm_names.csv')
```
